### 14. Horizontal Bar Plot: Song Duration Across Different Genres

```python
# Cleaning the id_artists from sp_tracks to match with the id from sp_artists
sp_tracks['id_artists_clean'] = sp_tracks['id_artists'].str.strip("[]'\"").str.split(',').str[0]
```

```python
merged_df = sp_tracks.merge(sp_artists[['id', 'genres']], left_on='id_artists_clean', right_on='id')
```

```python
merged_df.head()
```

```python
import ast

# Convert the string representation of lists to actual lists
merged_df['genres_list'] = merged_df['genres'].apply(lambda x: ast.literal_eval(x))

# Filter out rows where the genres_list is empty
filtered_genres_duration_df = merged_df[merged_df['genres_list'].apply(lambda x: len(x) > 0)]

# Explode the genres_list column to have one genre per row
exploded_genres_duration_df = filtered_genres_duration_df.explode('genres_list')

# Calculate the mean duration for each individual genre
genre_duration = exploded_genres_duration_df.groupby('genres_list')['duration'].mean().sort_values(ascending=False).reset_index()

top_genres = genre_duration.head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.set_palette('crest')

sns.barplot(y='genres_list', x='duration', data=top_genres)

plt.title('Average Duration of Songs by Genre')
plt.xlabel('Duration')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

### 15. Bar Plot: Top Five Genres by Popularity

```python
import ast

# Convert the string representation of lists to actual lists
merged_df['genres_list'] = merged_df['genres'].apply(lambda x: ast.literal_eval(x))

# Filter out rows where the genres_list is empty
filtered_genres_df = merged_df[merged_df['genres_list'].apply(lambda x: len(x) > 0)]

# Explode the genres_list column to have one genre per row
exploded_genres_df = filtered_genres_df.explode('genres_list')

# Calculate the mean popularity for each individual genre
genre_popularity = exploded_genres_df.groupby('genres_list')['popularity'].mean().sort_values(ascending=False).reset_index()

# Get the top 5 genres by popularity
top_5_genres = genre_popularity.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(y='genres_list', x='popularity', data=top_5_genres, palette='viridis')
plt.title('Top 5 Genres by Average Popularity')
plt.xlabel('Average Popularity')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```
